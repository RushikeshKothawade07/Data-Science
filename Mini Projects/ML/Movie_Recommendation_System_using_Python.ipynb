```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('/content/movies.csv')
```

```python
df.head()
```

```python
df.shape
```

```python

# selecting the relevant features for recommendation

selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#replacing the mising values with null strings
for feature in selected_features:
  df[feature] = df[feature].fillna('')
```

```python
df.head()
```

```python
#combining all the 5 selected cast
combined_features = df['genres'] + ' '+df['keywords'] + ' ' +df['tagline'] + ' ' +df['cast'] + ' '+ df['director']
```

```python
print(combined_features)
```

```python
#convertig text data into feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
#getting movie name from user
movie_name = input('Enter your favourite movie name :' )
#creating a list with all the movie names given in the dataset
list_of_all_titles = df['title'].tolist()
print(list_of_all_titles)
#finding the close match for the movie name given by user
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
print('------------------------------------------------------------------')
print('Recommended Movies : =>' )
print('------------------------------------------------------------------')

print(find_close_match)
close_match = find_close_match[0]
print('------------------------------------------------------------------')

print(close_match)
```

```python
#find the index of the movie with title
index_of_movie = df[df.title == close_match]['index'].values[0]
print(index_of_movie)
```

```python
#getting a list of similar movies
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score,key = lambda x : x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#print top 10 similar movies based on index
print('Movies suggested for you : ')
i=1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.index == index]['title'].values[0]
  if(i<=10):
    print(i,'.',title_from_index)
    i+=1
```
