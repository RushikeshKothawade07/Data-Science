##Vectors operations 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Dot product

```python
a = np.array([2,3])
b = np.array([4,4])
dot_prod  = np.dot(a,b)
print(dot_prod)
```

```python
c = np.array([20,30,55])
d = np.array([33,44,66])
dot_prod2 = np.dot(c,d)
print(dot_prod2)
```

```python
#dot product of two vectors is scaler
```

Cross Product

```python
a = np.array([2,3])
b = np.array([4,4])
cross_prod  = np.cross(a,b)
print(cross_prod)
```

```python
c = np.array([20,30,55])
d = np.array([33,44,66])
cross_prod2 = np.cross(c,d)
print(cross_prod2)
```

```python
x = np.array([2,5])
d = np.array([-8,6])

#magnitude of b vector
magnitude_of_d = np.sqrt(sum(d**2))

proj_of_x_on_d = (np.dot(x,d)/magnitude_of_d**2)*d
print(proj_of_x_on_d)
```

#MATRICES

```python
import numpy as np
mat1 = np.array([[2,3],[5,4]])
print(mat1)
print(mat1.shape)
```

```python
mat2 = np.array([[5,4,5],[2,1,2],[5,8,9],[5,6,4],[5,5,5]])
print(mat2)
print(mat2.shape)
```

```python
#creating matrices with random value
rmat= np.random.randint(100,size = (4,5))
print(rmat)
```

```python
#matrix with all values as 1
m3 = np.ones((2,3),dtype=int)
print(m3)
```

```python
#matrix with all values as 0 NULL/ZERO
m4 = np.zeros((4,4),dtype=int)
print(m4)
```

```python
#identity matrix
m5 = np.eye(2,3,dtype = int)
print(m5)
```

```python
#trnaspose of matrix
a = np.random.randint(100,size = (4,5))
print(a)
#trnaspose
b = np.transpose(a)
print(b)
```

Matrix operations

```python
m1 = np.random.randint(10,size=(5,5))
m2 = np.random.randint(10,size=(5,5))
sum = np.add(m1,m2)    #can also do m1+m2
diff = np.subtract(m1,m2) #can also do m1-m2
mul = np.multiply(m1,m2) #can also do m1*m2
prod = np.dot(m1,m2)
print('Matrix A ')
print(m1)
print('Matrix B ')
print(m2)
print('Addition is ')
print(sum)
print('Subtraction is ')
print(diff)
print('Multiply is ')
print(mul)
print('Dot product is ')
print(prod)
print(prod.shape)
```
