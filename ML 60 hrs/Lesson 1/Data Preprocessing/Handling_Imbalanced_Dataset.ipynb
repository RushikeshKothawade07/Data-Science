This is highly imbalanced dataset

```python
#separating the legit and fraudulent transactions
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape,fraud.shape)
```

Undersampling 

Build a sample dataset containing similar distribution of Legit and Fraudelant Transactions

```python
# No of fraudelant transactions = 362
```

```python
legit_sample = legit.sample(n=362)
```

```python
legit_sample.shape
```

Concatenate the two dataframes

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```
