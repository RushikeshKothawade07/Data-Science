#Extracting data

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content,'html.parser')
soup.title
soup.find_all('p')
```

```python
for para in soup.find_all('p'):
  print(para.text)
```

```python
for link in soup.find_all('a'):
  print(link.get('href'))
```

```python
for content in soup.find_all('div',id='toc',class_='tov'):
  print(content.text)
```

```python
rows_=[]
for table in soup.find_all('div',id='covid19-container'):
  for rows in table.find_all('tr'):
    rows_.append(rows.text)
```

```python
modified_rows = []
for row in rows_:
  modified_rows.append(list(filter(bool,row.splitlines())))
```

```python
import pandas as pd
df = pd.DataFrame(data = modified_rows )
```
